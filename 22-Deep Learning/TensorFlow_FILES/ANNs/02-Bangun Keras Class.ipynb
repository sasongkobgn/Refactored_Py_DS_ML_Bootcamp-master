Multisurface Method-Tree (MSM-T) [K. P. Bennett, "Decision Tree

programming to construct a decision tree.  Relevant features
were selected using an exhaustive search in the space of 1-4
features and 1-3 separating planes.

## Train Test Split
biasa `train_test_split` menggunaakan `model_selections`

```python
X = df.drop('benign_0__mal_1',axis=1).values
y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

## Scaling Data 


```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

ini adalah tanda function nya bagus

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# import tensorflow 
import tensorflow as tf 
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense, Activation, Dropout
```

### How to choose the number of hidden layers and nodes in a feedforward neural network?

```python
model = Sequential()

model.add(Dense(units=30,activation='relu'))

model.add(Dense(units=15,activation='relu'))

model.add(Dense(units=1,activation='sigmoid'))

# for binary classification problem
model.compile(loss='binary_crossentropy', optimizer='adam')
```

## Training the Model

## Example Two: Early Stopping

We obviously trained too much! Let's use early stopping to track the val_loss and stop training once it begins increasing too much!

kita mengguakan early stoping for this. biar ketika ada penaikan validation loss maka dia bisa stop sendiri

```python
model = Sequential()
model.add(Dense(units=30,activation='relu'))
model.add(Dense(units=15,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
# import keras callbacks Early stopping
from tensorflow.keras.callbacks import EarlyStopping
```

Stop training when a monitored quantity has stopped improving.

    Arguments:
        monitor: Quantity to be monitored.
        min_delta: Minimum change in the monitored quantity
            to qualify as an improvement, i.e. an absolute
            change of less than min_delta, will count as no
            improvement.
        patience: Number of epochs with no improvement
            after which training will be stopped.
        verbose: verbosity mode.
        mode: One of `{"auto", "min", "max"}`. In `min` mode,
            training will stop when the quantity
            monitored has stopped decreasing; in `max`
            mode it will stop when the quantity
            monitored has stopped increasing; in `auto`
            mode, the direction is automatically inferred
            from the name of the monitored quantity.
            
            
 

jadi gue pengent tahu gimana ini caranya, 
jadi diatas duah pernnah bialnag kalau. 
min_delta: perubahan minimum yang diamati dari kuantitas
patience: nilai epochs yang tidak berubah
verbosity: verbose mode, kesalahan mungkin ya


```python
# add earlyStopping to our model evaluation 
early_stop = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=25)
```

```python
# sekarang kita training lagi, siapa tahu ada perubahan. coba selela
model.fit(x=X_train, 
          y=y_train,
          epochs=600, 
          validation_data=(X_test,y_test), verbose=1,
          callbacks = [early_stop]
         )
```

```python
# print the model loss again for this model
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

nah sekarang ada sedikit perubahan, jadi validation loss tidak sebesar sebelumnya. mari kita bandingkan

## Example Three: Adding in DropOut Layers

kita bisa mengantisipasi dengan cara menambah atau mengurangi layers dari NN kita

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()
model.add(Dense(units=30,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=15,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(units=1,activation='sigmoid')) # ini gak di dropout
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model.fit(x=X_train, 
          y=y_train, 
          epochs=600,
          validation_data=(X_test, y_test), verbose=1,
          callbacks=[early_stop]
          )
```

```python
# add the history
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

terlihat perbedaan dari gambar diatas. validation loss hasilnya lebih kecil, mengapa seperti ini? 

sebenarnya kita tidak tahu bagaimana data tersebut seperti apa, kita ahanya menjalankan NN kita tanapa ktia tahu bagaimana hala tersebut. hanya mencari loss yang paling kecil bagaimana bisa di approach

```python
# sekarang kita melakukan predictions
from sklearn.metrics import classification_report,confusion_matrix
```

```python
predictions = model.predict_classes(X_test)
```

this is berhubungan dengan kesalahan-kesalahan yang ada pada pengukuran, jadi ada kesalahan tipe I dan tipe II, 

intinya sebuah recall yang bagus adalah yang hasilnya kecil,

_kalau dari wikipediannya jadi seperti ini,
precision (also called positive predictive value) is the fraction of relevant instances among the retrieved instances, while recall (also known as sensitivity) is the fraction of the total amount of relevant instances that were actually retrieved. Both precision and recall are therefore based on an understanding and measure of relevance_

```python
print(classification_report(y_test,predictions))
```

### Note: 
akurasi, dan presisi itu berbeda, jadi kita langsung bisa mendapat hasilnya dari sini. tanpa mengukur,

semakin banyak diulang akan semakin bagus, 

```python
print(confusion_matrix(y_test,predictions))
```
