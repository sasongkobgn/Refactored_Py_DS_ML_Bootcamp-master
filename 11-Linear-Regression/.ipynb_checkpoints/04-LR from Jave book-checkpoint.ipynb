```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns; sns.set()
%matplotlib inline
```

```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
plt.scatter(x, y);
```

selanjutnya kita akan membuat sebuah linear regression, yaitu seperti ini. 

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True) # model initialization

# model.fit = fitting linear dari data yang telah kita buat diatas
model.fit(x[:, np.newaxis], y)

# axis dan ordinat yang kita peroleh dari gambar
xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

# ploting scatter dari x and y:
plt.scatter(x, y)
plt.plot(xfit, yfit);
```

```python
# print model slope:
print("model_slope:      ", model.coef_[0])
print("model_intercept:  ", model.intercept_)
```

### Gausian Basis Fucntion

Of course, ini adalah fucntion yang kita gunakan, jadni terdapat banyak sekali fucntion, tidak hanya linear regreession saja namun bisa dengan function yang lain juga. tidak hanya berbentuk linear, kita berangkat dari sini. setelah kita menemukan funciton nya maka selanjutnya kita dapat membuat sebuah prediksi. 



```python
from sklearn.base import BaseEstimator, TransformerMixin

class GaussianFeatures(BaseEstimator, TransformerMixin):
    def __init__(self, N, width_factor=2.0):
        self.N = N
        self.width_factor = width_factor

    @staticmethod
    def _gauss_basis(x, y, width, axis=None): # ini adalah variable yang bakal kita pakai
        arg = (x - y) / width
        return np.exp(-0.5 * np.sum(arg ** 2, axis))

    def fit(self, X, y=None):
        self.centers_ = np.linspace(X.min(), X.max(), self.N)
        self.width_ = self.width_factor * (self.centers_[1] - self.centers_[0])
        return self

    def transform(self, X):
        return self._gauss_basis(X[:, :, np.newaxis], self.centers_, self.width_, axis=1)  

gauss_model = make_pipeline(GaussianFeatures(20), LinearRegression())

gauss_model.fit(x[:, np.newaxis], y)
yfit = gauss_model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit)
plt.xlim(0, 10); # rentang x(0, 10)
```

dari sini kita bisa mengerti bawasannya gaussian basis yang kita punya dapat diguakan untuk melinear-regression-kan beberapa bentuk scatter dari data. dari situ dapat disimpulakn bahwa kita bisa menggukaan berbagi function yang kita anggap sesuati atau cocok dengan function yang kita punya.
